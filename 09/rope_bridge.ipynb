```julia
mutable struct RopeSection
	x::Int
	y::Int
	head::Union{RopeSection,Nothing}
	final::Bool
	id::Any
end

visited = Set(["0,0"])

function moveHead(head::RopeSection, direction::String)
	if direction === "R"
		head.x += 1
	end
	if direction === "L"
		head.x -= 1
	end
	if direction === "U"
		head.y += 1
	end
	if direction === "D"
		head.y -= 1
	end
end

function markTail(tail, visited)
	position = "$(tail.x),$(tail.y)"
	push!(visited, position)
end

function moveSection(section::RopeSection, direction)
	if section.head === nothing
		moveHead(section, direction)
		return
	end

	Δx = section.head.x - section.x
	Δy = section.head.y - section.y

	if abs(Δx) > 1 && abs(Δy) > 1
		section.x += Δx÷2
		section.y += Δy÷2
	else
		if abs(Δx)> 1
			if Δx > 1
				section.x += 1
				section.y = section.head.y
			end
			if Δx < -1
				section.x -= 1
				section.y = section.head.y
			end
		end
	
		if abs(Δy)> 1
			if Δy > 1
				section.y += 1
				section.x = section.head.x
			end
			if Δy < -1
				section.y -= 1
				section.x = section.head.x
			end
		end
	end


	if section.final
		markTail(section, visited)
	end
	
end

function move(instruction,rope)
	direction, distance = instruction

	for i in 1:distance
		for section in rope
			moveSection(section,direction)
		end
		# println()
		# for i in 1:length(rope)
		# 	section = rope[i]
		# 	println("$(section.id): ($(section.x),$(section.y))")
		# end

	end
end

function generateRope(size)
	head = RopeSection(0, 0, nothing, false, "H")
	rope = [head]

	previous = head

	for i = 2:size-1
		section = RopeSection(0, 0, previous, false, "$(i-1)")
		push!(rope, section)
		previous = section
	end
	
	tail = RopeSection(0, 0, previous, true, "$(size-1)")
	push!(rope, tail)
	return rope
end

function getVisitedCount(data)

	rope = generateRope(10)

	for instruction in data
		move(instruction,rope)
	end


	return length(visited)
end

getVisitedCount(data)

```
